```python
import numpy as np

def KL(a, b):
    a = np.asarray(a, dtype=np.float)
    b = np.asarray(b, dtype=np.float)

    return np.sum(np.where(a != 0, a * np.log(a / b), 0))
```

```python
a=['workclass','education','occupation','race','sex','native_country']
#a=['workclass']
```

```python
f=['sum','count','min','max','avg']
#f=['sum']
```

```python
target=[]
ref=[]
app=[]
apps=[]
query=[]
for i in a :
    for j in m:
        for k in f:
            query.append([i,k,j])
            cur.execute("""select %s,%s(%s) from df where marital_status='Married' group by %s """ % (i,k,j,i))
            trows = cur.fetchall()
            #print (trows)
            conn.rollback()
            cur.execute("""select %s,%s(%s) from df where marital_status='Unmarried' group by %s  """ % (i,k,j,i))
            rrows = cur.fetchall()
            #print (rrows)
            
            
            norm=[]
            
            #print (i,k,j)
            for x,y in trows:
                app.append(int(y))
            
           
            
            for x,y in trows:
                norm.extend([((int(y)-min(app))/(max(app)-min(app)))])
            
            
            target.append(norm)
            
            
            
            #print (i,k,j)
            for x,y in rrows:
                apps.append(int(y))
            norm=[]
            for x,y in rrows:
                norm.extend([((int(y)-min(apps))/(max(apps)-min(apps)))])
            
            ref.append(norm)
            
```

## Query 1 country -max(loss)

```python
cur.execute("""select Distinct(native_country),max(capital_loss) from df where marital_status='Married' group by native_country """)
t = cur.fetchall()
cur.execute("""select Distinct(native_country),max(capital_loss) from df where marital_status='Unmarried' group by native_country """)
r = cur.fetchall()
fd=[]
for x,y in t:
    for a,b in r:
        if x==a:
            fd.append([a,int(y),int(b)])
xa=[]
for x,y,z in fd:
    if y!=0 and z!=0:
        xa.append([x,y,z])
print(xa)
```

```python
df = pd.DataFrame(xa)

ax = df.plot(kind='bar')
ax.set_xticks(df.index)
ax.set_xticklabels(df[0], rotation=90)
ax.legend(['Married','Unmarried'])
ax.set_xlabel("Native Country")
ax.set_ylabel("Maximum Capital Loss")
plt.tight_layout()
plt.savefig("Query 1.png")
```
